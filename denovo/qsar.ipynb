```python
from qsprpred.data.sources.papyrus import Papyrus

acc_keys = ["P08908"]  # accession keys of the proteins to fetch data for
name = "P08908_LIGANDS_nostereo"  # name of the data set
quality = "high"  # choose minimum quality from {"high", "medium", "low"}

# fetch data from Papyrus for the given targets (may take a while)
papyrus = Papyrus(
    data_dir="./data/data",  # where to store the data
    stereo=False,  # do not consider stereochemistry
    plus_only=True,  # only use the high quality data set
    descriptors=None,  # do not download descriptors (we will calculate them later)
)
dataset = papyrus.getData(
    name,
    acc_keys,
    quality,
    use_existing=True # use existing data set if it was already compiled before
)
dataset.getDF()
```

```python
dataset.getProperties()
```

```python
dataset.getSubset("type_")
```

```python
import numpy as np

def parse_values(row):
    mask = np.array([x == "1" for x in row["type_Ki"].split(";")], dtype=bool)
    if sum(mask) == 0:
        return np.nan
    vals = np.array(row["pchembl_value"].split(";"), dtype=float)
    if len(mask) == len(vals):
        return np.median(vals[mask])
    else:
        return np.median(vals)

parsed_vals = dataset.getDF().apply(parse_values, axis=1)
dataset.addProperty("pKi_median", parsed_vals)
dataset.getProperty("pKi_median")
```

```python
dataset.getProperty("pKi_median").isna().sum()
```

```python
to_keep = dataset.getProperty(dataset.idProp)[~dataset.getProperty("pKi_median").isna()]
to_keep
```

```python
from qsprpred.data import QSPRDataset

dataset = QSPRDataset(
    df=dataset.getDF().loc[to_keep], 
    name=dataset.name, 
    index_cols=[dataset.idProp],
    target_props=[{"name": "pKi_median", "task": "REGRESSION"}],
    random_state=42
)
len(dataset), dataset.getProperty("pKi_median").isna().sum()
```

```python
from qsprpred.data.descriptors.fingerprints import MorganFP
from qsprpred.data.descriptors.sets import RDKitDescs
from qsprpred.data import RandomSplit, ClusterSplit

# Specifiy random split for creating the train (80%) and test set (20%)
split = RandomSplit(test_fraction=0.2)
# split = ClusterSplit(test_fraction=0.2)

# calculate compound features and split dataset into train and test
dataset.prepareDataset(
    split=split,
    feature_calculators=[MorganFP(radius=3, nBits=2048), RDKitDescs()],
)

print(f"Number of samples in train set: {len(dataset.y)}")
print(f"Number of samples in test set: {len(dataset.y_ind)}")
```

```python
from qsprpred.models import SklearnModel
from xgboost import XGBRegressor
from qsprpred.models import CrossValAssessor, TestSetAssessor
from qsprpred.models import OptunaOptimization

# This is an SKlearn model, so we will initialize it with the SklearnModel class
model = SklearnModel(
    base_dir="data/models/qsar",
    alg=XGBRegressor,
    name=f"{dataset.name}_XGBRegressor"
)
search_space = {
    'max_depth': ["int", 3, 10],
    'learning_rate': ["float", 0.001, 0.4],
    'subsample': ["float", 0.25, 1],
    "n_estimators": ["int", 100, 500],
    "min_child_weight": ["int", 1, 10]
}

gridsearcher = OptunaOptimization(
    n_trials=100,
    param_grid=search_space,
    model_assessor=CrossValAssessor(scoring='r2'),
)
gridsearcher.optimize(model, dataset, refit_optimal=True)

# We can now assess the model performance on the training set using cross validation
CrossValAssessor("r2")(model, dataset)

# and on the test set
TestSetAssessor("r2")(model, dataset)

# Finally, we need to fit the model on the complete dataset if we want to use it further
# This will save the fitted model and metadata to disk
# model.fitDataset(dataset)

# We can optionally save the model and metadata to disk explicitly as well
_ = model.save()
```
